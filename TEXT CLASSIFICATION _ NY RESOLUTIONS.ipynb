```python
from textblob import TextBlob
from textblob.classifiers import NaiveBayesClassifier
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.metrics import accuracy_score
import csv
import nltk
import pandas
import re

trainset = pandas.read_csv('nyresolutions.csv', usecols=["text", "Resolution_Category"])
testset = pandas.read_csv('test.csv', usecols=["text", "Resolution_Category"])
textset = trainset["text"].tolist()
cat = trainset["Resolution_Category"].tolist()
tweet_test = testset["text"].tolist()
cat_test = testset["Resolution_Category"].tolist()
```

```python
def cleaning_data(data_tweet): 
        text = ' '.join(re.sub(r'@[A-Za-z0-9]+','',data_tweet).split()) 
        text = ' '.join(re.sub('https?://[A-Za-z0-9./]+','',text).split())
        text = ' '.join(re.sub("[^a-zA-Z]", " ", text).split())
        text = ' '.join(re.sub("RT", " ", text).split())
        text = ' '.join(re.sub(' +',' ', text).split())
        return text

cleaning = []
a = 0
while a < len(textset):
    cleaned = cleaning_data(textset[a])
    cleaning.append(cleaned)
    a += 1
    
```

```python
def word_tokenization(s):
    tokens = word_tokenize(s)

    return tokens

stop_words = set(stopwords.words('english')) 
    
def stopword(str):
    word_tokens = word_tokenize(str) 
    filtered_sentence = [w for w in word_tokens if not w in stop_words]  
    filtered_sentence = [] 
    
    for w in word_tokens: 
        if w not in stop_words: 
            filtered_sentence.append(w)
    return filtered_sentence

stopword_data = []
a = 0
while a < len(textset):
    new = ' '.join(stopword(cleaning[a]))
    stopword_data.append(new)
    a += 1
    

train_set = []
b=0
while b < len(textset):
    set_data = (stopword_data[b],cat[b])
    train_set.append(set_data)
    b += 1

print(train_set)
```

```python
cl = NaiveBayesClassifier(train_set)
#cl.classify("I feel amazing!")
blob = TextBlob("For Real The Struggle damn true", classifier=cl)
print(blob.classify())
```

```python
n = 0
while n < len(tweet_test):
    blob = TextBlob(cleaning[n], classifier=cl)
    print(blob.classify() + " "+"|"+" " + cat_test[n])
    n += 1
```
